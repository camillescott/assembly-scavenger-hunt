```python
import binascii

bin_to_dna_table = { '00':'A', '0':'A', '1':'A', '01':'G', '10':'C', '11':'T' }
dna_to_bin_table = { 'A':'00', 'G':'01', 'C':'10', 'T':'11' }

def text_to_bin(text):
    tmp = bin(int(binascii.hexlify(text),16))
    return tmp

def bin_to_text(bi):
    i = int(bi, 2)
    return binascii.unhexlify('%x' % i)

def bin_to_dna(bi):
    dna = ''
    for i in range(2,len(bi),2):
        key = bi[i:i+2]
        dna += bin_to_dna_table[key]
    return dna

def dna_to_bin(dna):
	bi = ''
	for n in dna:
		bi += dna_to_bin_table[n]
	return bi
```

```python
def random_seq(length):
    import random
    random.seed(1)

    x = ["A"] + ["G"] + ["C"] + ["T"]
    x = x*length
    random.shuffle(x)
    x = "".join(x)

    return x
```
